```python
import pandas as pd
```

```python
data = pd.read_csv('video_games_sales.csv')
pd.to_numeric(data['Critic_Score'])
data = data[data['Critic_Score'].notna()]
data.head()
```

####  Как критики относятся к спортивным играм?
H0: Средняя оценка спортивной игры критиками >= 80%  
H1: Средняя оценка спортивной игры критиками меньше 80%

За alpha будем принимать 0.05


```python
alpha = 0.05
```

```python
sport_games = data.loc[data['Genre'] == 'Sports']
sport_games.head()
```

```python
from scipy import stats as st
result = st.ttest_1samp(sport_games['Critic_Score'], 7.5)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее меньше 7.5')
else:
    print('Не отвергаем нулевую нулевую гипотезу, спортивные игры нравятся критикам')
```

#### Критикам нравятся больше игры на PC или на PS4?

H0: Средняя оценка игр на PC и PS4 одинакова   
H1: Средняя оценка игр на PC и PS4 различается 

За alpha будем принимать 0.05

```python
PC_games = data.loc[data['Platform'] == 'PC']
PS4_games = data.loc[data['Platform'] == 'PS4']
```

```python
result = st.ttest_ind(PC_games['Critic_Score'], PS4_games['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к играм PC и PS4')
    if result.statistic > 0:
        print("Критики любят игры на PC больше, чем на PS4")
    else:
        print("Критики любят игры на PS4 больше, чем на PC")
else:
    print('Не отвергаем нулевую гипотезу, игры нравятся одинаково')
```

#### Критикам больше нравятся стрелялки или стратегии?

H0: Средняя оценка стрелялок и стратегий одинакова  
H1: Средняя оценка стрелялок и стратегий различается  

За alpha будем принимать 0.05

```python
shooter_games = data[data.Genre == 'Shooter']
strategy_games = data[data.Genre == 'Strategy']
```

```python
result = st.ttest_ind(shooter_games['Critic_Score'], strategy_games['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к играм')
    if result.statistic > 0:
        print("Критики любят стрелялки больше, чем стратегии")
    else:
        print("Критики любят стрелялки больше, чем стратегии")
else:
    print('Не отвергаем нулевую гипотезу, игры нравятся одинаково')
```

```python
spam = pd.read_csv('spam.csv')
spam.head()
```

```python
#1) Привидите весь текст к нижнему регистру;

spam['Message'] = spam['Message'].str.lower()
spam.head()
```

```python
import re
```

```python
#2) Удалите мусорные символы;

spam_words = spam
spam_words['Message'] = spam_words['Message'].apply(lambda x: re.sub('[\W_]+',' ', x))
spam_words.head()
```

```python
#3) Удалите стоп-слова;
spam_words['Message'] = spam_words['Message'].apply(lambda x: x.split())
spam_words.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords_set = set(stopwords.words('english'))
without_sw = spam_words
```

```python
without_sw['Message'] = without_sw['Message'].apply(lambda x: [word for word in x if word not in stopwords_set])
without_sw.head()
```

```python
#4) Привидите все слова к нормальной форме;
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
spam_lemmatized = without_sw
```

```python
spam_lemmatized['Message'] = spam_lemmatized['Message'].apply(lambda x:[wordnet_lemmatizer.lemmatize(word) for word in x])
```

```python
spam_lemmatized.head()
```

```python
# 5) Преобразуйте все сообщения в вектора TF-IDF.
spam_lemmatized['Message'] = spam_lemmatized['Message'].apply(lambda x: ' '.join(x))
spam_lemmatized.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam_lemmatized['Message'])
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
#Разделите данные на тестовые и тренировочные в соотношении 30/70, укажите random_state=42. Используйте train_test_split;
from sklearn.model_selection import train_test_split
```

```python
tfidf_matrix['Category'] = spam['Category']
tfidf_matrix['Message'] = spam['Message']
tfidf_matrix.head()
```

```python
spam_train, spam_test = train_test_split(tfidf_matrix, test_size=0.3, random_state=42)
```

```python
#7) Постройте модель логистической регрессии, укажите random_state=42, оцените ее точность на тестовых данных;
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=42)

y_train = spam_train['Category']
y_test = spam_test[['Category', 'Message']]

del spam_train['Category']
del spam_test['Category']
del spam_train['Message']
del spam_test['Message']
```

```python
lr = lr.fit(spam_train, y_train)
lr.predict(spam_test)
```

```python
result = pd.DataFrame([y_test['Category'].values, lr.predict(spam_test)]).T
result.head()
```

```python
# 8) Опишите результаты при помощи confusion_matrix;
from sklearn.metrics import confusion_matrix
```

```python
matrix = confusion_matrix(y_test['Category'].values, lr.predict(spam_test))
print(matrix)
print('Было предсказано', matrix[0][0] + matrix[1][1], 'результатов а это:', (matrix[0][0] + matrix[1][1])/len(result.index)*100,'%')
```
